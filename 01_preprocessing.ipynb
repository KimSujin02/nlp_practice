```python
import pandas as pd

wiki = pd.read_csv("./namuwikitext_20200302.dev", sep="\t", header=None)
wiki.head()
```

```python
# 한글과 공백을 제외한 모든 문자를 제거
wiki[0] = wiki[0].replace(r'[^가-힣ㄱ-ㅎㅏ-ㅣ\s]', '', regex=True)
```

```python
from konlpy.tag import Okt
from tqdm import tqdm

okt = Okt() # 형태소 분석기 객체 생성

# 불용어 파일 생성 및 이용
with open('ko_stopwords.txt', 'r', encoding='UTF-8') as f:
    ko_stopwords =  [line.strip() for line in f]

preprocess_data = []
# 형태소 분석 및 불용어 제거
for sentence in tqdm(wiki[0]) :   
    tokens = okt.morphs(sentence, stem=True)
    tokens = [token for token in tokens if token not in ko_stopwords]
    preprocess_data.append(tokens)
```

```python
import pickle

with open("preprocess_data.pkl", "wb") as f:
    pickle.dump(preprocess_data, f)
```
