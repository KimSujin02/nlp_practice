# BoW 기반 텍스트 분석 연습

1. 적절한 데이터셋을 찾거나 생성하고, 적절한 전처리를 진행한다. [01_preprocessing.ipynb]
2. TF-IDF Vectorizer를 이용하여 벡터화한다.
3. Cosine Similarity를 계산하여 입력된 문자열의 긍/부정을 판단한다.

```python
import pickle

with open("./preprocessed_data/preprocess_data.pkl", "rb") as f:
    preprocess_data = pickle.load(f)
```

```python
documents = [" ".join(tokens) for tokens in preprocess_data]
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer(min_df=2)
text_vecs = tfidf_vectorizer.fit_transform(documents)
vocab = tfidf_vectorizer.get_feature_names_out()
print(text_vecs.shape)
cosine_similarity(text_vecs[1:2], text_vecs)
```
