# Word Embedding을 활용한 단어 찾기 게임

- Level 1. [Semantle](https://semantle.com/) 흉내내기
- Level 2. Semantle의 한국어 버전을 만들어보기
- Level 3. 단어 `A:B=C:?` 여기서 ? 맞추기 ([한국어 임베딩 확인](https://word2vec.kr/search/) + Semantle)

```python
import pickle

with open("./preprocessed_data/preprocess_data.pkl", "rb") as f:
    preprocess_data = pickle.load(f)
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=preprocess_data,
    vector_size=100,
    window=5,
    min_count=5,
    sg=1
)
```

```python
model.wv.vectors.shape
```

```python
import random

target_word = random.choice(preprocess_data[1])
# print(target_word)

repeat_num = input('반복 횟수를 입력하세요:')

similarity_list = []
for i in range(int(repeat_num)):
    input_word = input('단어를 입력하세요:')

    similarity = model.wv.similarity(target_word, input_word)
    print(f"유사도 : {similarity:.2f}")
    
    similarity_list.append(similarity)
    if i == int(repeat_num) - 1:
        print("\n게임 종료!")
        print(f"정답 : {target_word}")
        print(f"최고 유사도 : {max(similarity_list):.2f}")
```
